```python
metadata=dict()
data=[]
indexlist=[]

for idtje in range(1, 964):
    epubid = 'ebook'+str(idtje)
    ff='ePubs/results/'+epubid+'/report.json'
    try:
        with open(ff, encoding="utf8") as f:
            d = json.load(f)
            metadata[epubid]=d['earl:testSubject']['metadata']
            A=d['assertions']
            errors=[]
            for d in A:
                if d['earl:result']['earl:outcome']=='fail':
                    for a in d['assertions']:
                        if a['earl:result']['earl:outcome']=='fail':
                            errors.append(a['earl:test'])
                            indexlist.append(epubid)
                     
            data=data+errors
    except:
        pass
# make dataframes        
metadf=pd.DataFrame.from_dict(metadata, orient='index')
errordf=pd.DataFrame(data)
errordf['epubID']=indexlist
```
