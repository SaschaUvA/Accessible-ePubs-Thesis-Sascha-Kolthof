```python
metadata=dict()
data=[]
indexlist=[]

for epubid in list(df.index):
    ff='ePubs/results/'+epubid+'/report.json'
    try:
        with open(ff, encoding="utf8") as f:
            d = json.load(f)
            metadata[epubid]=d['earl:testSubject']['metadata']
            A=d['assertions']
            errors=[]
            for d in A:
                if d['earl:result']['earl:outcome']=='fail':
                    for a in d['assertions']:
                        if a['earl:result']['earl:outcome']=='fail':
                            errors.append(a['earl:test'])
                            indexlist.append(epubid)
                     
            data=data+errors
    except:
        pass
# make dataframes        
metadf=pd.DataFrame.from_dict(metadata, orient='index')
errordf=pd.DataFrame(data)
errordf['epubID']=indexlist
```

#  aantal en welke fouten per uitgever

```python
medf=pd.merge(errordf,metadf, left_on='epubID', right_index=True)
medf.head()
```

```python
medf2=medf.dropna(subset=['dc:publisher'])
medf2['dc:publisher'] = medf2['dc:publisher'].astype(str)
```

```python
ErrorsPerPublisher=medf2.groupby('dc:publisher')['earl:impact'].value_counts().unstack().fillna(0).astype(int)
ErrorsPerPublisher
```
